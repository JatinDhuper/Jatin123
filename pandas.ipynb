```python
#Question 1
import pandas as pd

# Create the Series
data = pd.Series([100, 200, 300, 400], index=['x', 'y', 'z', 'w'])

# Access the value at index 'y'
value_y = data.loc['y']

print("The value at index 'y' is:", value_y)
```

```python
#Question 2
# Access the 2nd element using .iloc[] 
second_element = data.iloc[1] 
print("The 2nd element in the Series is:", second_element)
```

```python
#Question 3
# Access multiple elements using .loc[] 
selected_values = data.loc[['x', 'w']] 
print("Values at indices 'x' and 'w':") 
print(selected_values)
```

```python
#Question 4
# Access the last element using .iloc[] 
last_element = data.iloc[-1] 
print("The last element in the Series is:", last_element)
```

```python
#Question 5
# Access elements from index 'y' to 'z' (inclusive) using .loc[] 
subset = data.loc['y':'z'] 
print("Elements from index 'y' to 'z':") 
print(subset)
```

```python
#Question 6
import pandas as pd 
# Series with integer-based indices 
data = pd.Series([10, 20, 30, 40], index=[0, 1, 2, 3]) 
# Using .iloc[2] -> position-based 
print("Using .iloc[2]:", data.iloc[2]) 
# Using .loc[2] -> label-based 
print("Using .loc[2]:", data.loc[2])
```

```python
#Question 7
# Access the element with index 'x' using dot notation 
value_x = data.x 
print("The value at index 'x' is:", value_x)
```
